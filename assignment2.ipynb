# Importing Data

DATA INFO

####**Merging all the data frames by city**
Putting all 4 tab/dataframes together 

```python
#MERGE BY 'CITY' columns
merged_df = df1.merge(df2, how='inner', on = 'CITY').merge(df3,  how='inner', on='CITY').merge(df4,  how='inner', on='CITY')
merged_df

```

```python
#veryfying using locate Aracaju to make sure it is not dropped
merged_df.loc[merged_df['CITY'].str.contains("Aracaju", case=False)].head(2)
```

```python
#Originaly used this filtering to analyze the df for where the duplicate was occuring.
#Since we were getting 302 rows for brasillia after merge, we want to make sure deleting it from our df4 corrected this.
merged_df.loc[merged_df['CITY'].str.contains("Brasília", case=False)]
```

####Trying different ways of merging to see which one is best

Using the 'Interior/Metropolitan' column was best as merging using this column returned 4077 rows (Interpreted as no duplicates). 151 days of recorded data times 27 capital city would give us 4077 rows. The other code blocks were deleted.

```python
merge_via_interior = merged_df.loc[merged_df['Interior/Metropolitan'] == 1.0]
final_df = merge_via_interior.dropna(subset=['CITY'])
final_df
```

## **Creating our features dataframe**
Since the 'number of days since the outbreak' will be each row of recorded data for each capital city, we need an index column.

```python
final_df['days'] = final_df.groupby('CITY').cumcount()
#Confirming that it counted the 151 days for each city
view_index = final_df[['days', 'CITY']]
view_index
```

####Putting it together
The features DataFrame must contain the following feature columns in this exact order: days_cube,
days_sq, days, temp, pop_dense_sq, pop_dense, pop.

```python
#creating the data frame
features = pd.DataFrame()
features
```

#### **Getting our days variables**

```python
#cubing and squaring the days index/count column we created and puttiing them into our dataframe
features['days_cube']=final_df['days']**3
features['days_sq']=final_df['days']**2
features['days']=final_df['days']
features
```

####**Add Temp feature**

```python
#adding the ANNUAL column from the merged df and naming it
features['temp']=final_df['ANNUAL']
features
```

####**Add Popolation Features**
●pop_dense_sq is the squared value of pop_dense
● pop is the city population
● Population density is defined as the population of a city divided by the area in kilometers
squared of the city.

```python
#checking data type 
final_df.dtypes['Population as of 2019']
```

```python
final_df.dtypes['SQ_KM']
```

```python
#since we need to do calculations to the ['Population as of 2019'] column, we need to change it to an interger
final_df['Population as of 2019'] = final_df['Population as of 2019'].astype('int')
final_df.dtypes['Population as of 2019']
```

```python
#Doing our calculations to the population column and adding the needed columns to our df
#order:pop_dense_sq, pop_dense, pop
features['pop_dense_sq']=(final_df['Population as of 2019']/final_df['SQ_KM'])**2
features['pop_dense']=final_df['Population as of 2019']/final_df['SQ_KM']
features['pop']=final_df['Population as of 2019']
#viewing our completed features DF
features
```
